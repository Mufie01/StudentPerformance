```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Load dataset (upload CSV from Kaggle to your runtime or local)
df = pd.read_csv("StudentsPerformance.csv")
```

```python
# Step 3: Preview dataset
print(df.head())
print(df.info())
print(df.describe())
```

```python
# Step 4: Check for missing values
print(df.isnull().sum())
```

```python
# Step 5: Analyze score patterns
print(df.groupby('gender')[['math score', 'reading score', 'writing score']].mean())
print(df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean())
print(df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean())
```

```python
# Step 6: Visualization
sns.set(style="whitegrid")
# Math scores by gender
sns.barplot(x='gender', y='math score', data=df)
plt.title("Average Math Score by Gender")
plt.show()

# Reading scores by test preparation
sns.boxplot(x='test preparation course', y='reading score', data=df)
plt.title("Reading Score by Test Prep Course")
plt.show()

# Writing score distribution by lunch
sns.histplot(data=df, x='writing score', hue='lunch', bins=20, multiple='stack')
plt.title("Writing Score Distribution by Lunch Type")
plt.show()
```

```python
# Step 7: Insights
print("""
Key Insights:
1. Students who completed the test preparation course scored higher on average.
2. Standard lunch students perform better than those with free/reduced lunch.
3. Male students scored higher in math, while female students scored better in reading and writing.
4. Parental education level shows a slight correlation with student scores.
""")
```
